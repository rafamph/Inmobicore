```python
def missing_elements(L, start, end):
    if end - start <= 1: 
        if L[end] - L[start] > 1:
            yield from range(L[start] + 1, L[end])
        return

    index = start + (end - start) // 2

    # is the lower half consecutive?
    consecutive_low =  L[index] == L[start] + (index - start)
    if not consecutive_low:
        yield from missing_elements(L, start, index)

    # is the upper part consecutive?
    consecutive_high =  L[index] == L[end] - (end - index)
    if not consecutive_high:
        yield from missing_elements(L, index, end)

def main():
    L = [10,11,11,13,14,15,16,17,18,20]
    print(list(missing_elements(L,0,len(L)-1)) + L)
    L = range(10, 21)
    print(list(missing_elements(L,0,len(L)-1)))

main()
```

```python
preciospormetro = []
m = 0
historial = {}
for delegacion in list(result.keys()):
    keyID = list(result[list(result.keys())[m]].keys())[0]
    subresults = result[list(result.keys())[m]][keyID]
    preciospormetro = []
    for i in range(len(subresults)):
        try:
            metros = int(subresults[i]["Construidos"].replace("m²",""))
            precio = int(subresults[i]["precio"])
            if precio > 200000:
                preciospormetro.append(precio/metros)
        except KeyError:
            pass
    historial[delegacion] = {"mean": mean(preciospormetro), "stdev": stdev(preciospormetro)}

    print("Delegación: "+delegacion)
    print(str(mean(preciospormetro))+"\n")
    m += 1

```
